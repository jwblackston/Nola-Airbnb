```python
import pandas as pd
nola_list = pd.read_csv('listings.csv')

nola_list.describe()
```

```python
nola_list.head(5)
```

```python
#checking missingness as a percentage:
missing = nola_list.isnull().sum()/len(nola_list)*100
print(missing)
```

```python
#scrub columns/variaables for their names and eliminate greater than 10% missingness
features = nola_list.columns
feats = []

for f in range(len(features)):
    if missing[f] <= 10:
        feats.append(features[f])

print(feats)
print(len(feats))
```

```python
import numpy as np

acc_value = 3
first_living_space_value = nola_list.loc[0,'accommodates']
first_distance = np.abs(first_living_space_value - acc_value)

print(first_distance)

nola_list['distance'] = np.abs(nola_list.accommodates - acc_value)
nola_list.distance.value_counts().sort_index()

nola_list = nola_list.sample(frac=1,random_state=0)
nola_list = nola_list.sort_values('distance')
nola_list.price.head()
```

```python
drop_list = ['id', 'host_id', 'description', 'summary', 'experiences_offered', 'house_rules', 'thumbnail_url', 'medium_url', 'picture_url',
            'listing_url', 'scrape_id', 'last_scraped', 'name', 'summary', 'space', 'neighbourhood', 'neighbourhood_group_cleansed',
            'neighborhood_overview', 'host_about', 'calendar_last_scraped', 'host_has_profile_pic',
            'city', 'state', 'zipcode', 'market', 'smart_location', 'host_url', 'xl_picture_url', 'transit', 'notes', 'access',
             'host_name', 'host_location', 'interaction', 'host_response_time', 'host_thumbnail_url', 'host_picture_url',
             'jurisdiction_names', 'host_neighbourhood','host_listings_count','host_total_listings_count','host_verifications','host_identity_verified',
             'street', 'country_code', 'country', 'latitude', 'longitude', 'is_location_exact', 'amenities', 'weekly_price',
             'monthly_price', 'calendar_updated', 'host_since', 
            'license']

update_df = nola_list.drop(drop_list, axis = 1)
```

```python
#list vars that are 'object' type:
object_columns_df = update_df.select_dtypes(include=['object'])
print(object_columns_df.iloc[0])
```

```python
#remove trailing percentage:
update_df['host_response_rate'] = update_df['host_response_rate'].str.rstrip('%').astype('float')
```

```python
#explore categorical variables
cols = ['host_is_superhost', 'neighbourhood_cleansed', 'property_type', 'room_type', 
       'bed_type', 'has_availability', 'requires_license', 
        'is_business_travel_ready',
       'cancellation_policy', 'require_guest_profile_picture', 'require_guest_phone_verification']
for name in cols:
    print(name,':')
    print(object_columns_df[name].value_counts(),'\n')
    
```

```python
df_dummy = pd.get_dummies(update_df[cols])
#update_df = update_df.join(df_dummy)
#update_df.head(5)

df_new = pd.concat([update_df, df_dummy], axis=1)
df_new.head(5)
```

```python
print("Data types and their frequency\n{}".format(df_new.dtypes.value_counts()))
```

```python
#convert values that will not get processed in models correctly
replacements = {
      r't': 1,
      r'f': 0}

df_new.replace(replacements, regex=True, inplace=True)
print(df_new)
```

```python
df_new.to_csv('nola_airbnb_reduced.csv', index=False)
```
